```python
import glob
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset = sorted(glob.glob('data*.csv'))
print(dataset)
```

```python
# Import the 1st dataset
data1 = pd.read_csv('data-01.csv')

data1.head()
```

```python
Min1 = data1['Min'].sum()
Min1
```

```python
#Plot Lines
plt.plot(data1.Average, label='Average')
plt.plot(data1.Min, label='Min')
plt.plot(data1.Max, label='Max')

#Titles
plt.title('Dataset 1')
plt.xlabel('Feature#')
plt.ylabel('Value')

#Legend 
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()

#if statements
if min1==0:
    print("Why is this happening?")
    
elif 0==0 and 20==20:
    print("I doubt the integrity of this dataset") 
    
else:
    print("Dataset looks good")
```
